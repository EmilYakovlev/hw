```python
#Python 3
def quicksort(array):
    less = []
    equal = []
    greater = []
    if len(array) > 1:
        pivot = array[0]
        for x in array:
            if x < pivot:
                less.append(x)
            if x == pivot:
                equal.append(x)
            if x > pivot:
                greater.append(x)
        return quicksort(less) + equal + quicksort(greater) 
    else:  
        return array
a = [5, 2, 3, 1, 4]
a = quicksort(a)
print a
```

```python
#NumPy 1
import numpy as np
import math
def mylinspace(x, y, z):
    arr = [x]
    i = x
    while i < y:
        i = i + math.fabs(x-y) / (z-1)
        arr.append(i) 
    return arr 
def mylinspace1(x, y, z):
    res = [np.arange(x, y, math.fabs(x-y) / (z-1))]
    return res
print mylinspace(-3, 3, 13)
%timeit mylinspace(-3, 3, 13)
print mylinspace1(-3, 3, 13)
%timeit mylinspace1(-3, 3, 13)
print np.linspace(-3, 3, 13)
%timeit np.linspace(-3, 3, 13)
```

```python
#Matplotlib
%matplotlib inline
from matplotlib import pylab as plt
from matplotlib import gridspec
x = np.arange(-5, 5, 0.1)
fig = plt.figure(figsize=(10, 5)) 
gs = gridspec.GridSpec(1, 2) 
ax0 = plt.subplot(gs[0])
ax0.plot(x, x, label = 'y = x')
ax0.plot(x, x**2, label = 'y = x^2')
ax0.plot(x, x**3, label = 'y = x^3')
ax0.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y') 
ax1 = plt.subplot(gs[1])
ax1.plot(x, np.sin(x), label = 'y = sin x')
ax1.plot(x, np.cos(x), label = 'y = cos x')
ax1.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, mode="expand", borderaxespad=0.)
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('one.png')
plt.show()
```
